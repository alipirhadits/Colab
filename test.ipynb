```python
raw = "  Temp=  23.5 C ; RH=  61 %  "
```

```python
raw = raw.strip()
raw
```

```python
parts = raw.split(';')
temp_part = parts[0]
rh_part = parts[1]
```

```python
temp_value = ''.join(ch for ch in temp_part if ch.isdigit() or ch == ".")
rh_value = ''.join(ch for ch in rh_part if ch.isdigit())
```

```python
result = f"Temp: {temp_value}, RH: {rh_value}"
```

```python
print(result)
```
